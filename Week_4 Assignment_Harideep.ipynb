```python
import re

# Define a function to filter and clean numeric strings
def clean_numeric_column(column):
    # Use regex to keep only valid numeric strings (including commas)
    column_cleaned = column.astype(str).apply(lambda x: re.sub(r'[^\d.,-]', '', x))
    # Convert to numeric, coercing errors
    return pd.to_numeric(column_cleaned.str.replace(',', ''), errors='coerce')

# Apply the cleaning function to the relevant columns
df['annual_salary'] = clean_numeric_column(df['annual_salary'])
df['additional_compensation'] = clean_numeric_column(df['additional_compensation'])
```

```python
import pandas as pd
import pycountry
from fuzzywuzzy import process

# Get a dictionary of all official currency names mapped to their ISO 4217 codes
all_currency_names = {c.name: c.alpha_3 for c in pycountry.currencies}
print(all_currency_names)


# Function to standardize currency names
def get_currency_code(raw_name):
    raw_name = raw_name.strip()
         
    
    if raw_name in all_currency_names:
        return all_currency_names[raw_name]

    # Use fuzzy matching to find the closest valid currency name
    best_match, score = process.extractOne(raw_name, all_currency_names.keys())
    if score > 80:  # Confidence threshold
        return all_currency_names[best_match]
    
    for key, value in all_currency_names.items():
        if raw_name in value:
            return value
    
    return None

# Apply currency standardization to create a new '_currency' column
df['_currency_other'] = df['currency_other'].apply(lambda x: get_currency_code(x) if pd.notnull(x) else None)
```

```python
df[df.currency_other.notnull()][['_currency_other', 'currency_other']].sample(15)
```

```python
# read the currency conversion rates json file
conversion_rates = pd.read_json('conversion_rates.json')
exchange_rates_dict = conversion_rates.set_index('Currency').to_dict()['Exchange_Rate']
exchange_rates_dict
```

```python

sample_indexes = df[df.currency == 'Other'].sample(5).index
print("Bedore standardization:")
print(df.loc[sample_indexes, ['currency', 'currency_other', '_currency_other']])

df[df.currency == 'Other'] = df[df.currency == 'Other'].assign(currency=df['_currency_other'])

print("After standardization:")
print(df.loc[sample_indexes, ['currency', 'currency_other', '_currency_other']])
```

```python
def convert_to_usd(row, column_name='annual_salary'):
    # If the currency is USD, return the original amount
    if row['currency'] == 'USD':
        return row[column_name]
    
    # If the currency is not USD, look up the exchange rate
    exchange_rate = exchange_rates_dict.get(row['currency'])
    if exchange_rate is None:
        return None
    
    return row[column_name] / exchange_rate

# Apply the conversion to create a new 'annual_salary_usd' column
df['annual_salary_usd'] = df.apply(lambda row: convert_to_usd(row, 'annual_salary'), axis=1)
df['additional_compensation_usd'] = df.apply(lambda row: convert_to_usd(row, 'additional_compensation'), axis=1)

df[df.currency != 'USD'][['annual_salary', 'currency', 'annual_salary_usd', 'additional_compensation', 'additional_compensation_usd']].sample(5)
```

```python
df.us_state.fillna('Non US State', inplace=True)
df.city.fillna('Not specified', inplace=True)
df.isna().sum()
```

```python
df.education_level.fillna('Not specified', inplace=True)
print(df.education_level.value_counts())

education_levels = {
    "Not specified": 0,
    "College degree": 1,
    "Master's degree": 2,
    "Some college": 3,
    "PhD": 4,
    "Professional degree (MD, JD, etc.)": 5,
    "High School": 6
}

df.education_level = df.education_level.map(education_levels)
df.isna().sum()
```

```python
# Replacing Other or prefer not to answer, and null values with 'prefer not to answer'

df.gender.fillna('Prefer not to answer', inplace=True)
df.gender.replace({'Other or prefer not to answer': 'Prefer not to answer'}, inplace=True)

print(df.gender.value_counts())

df.isna().sum()
```

```python
df.race.fillna('Another option not listed here or prefer not to answer', inplace=True)

# Define the master list of races
race_master_list = [
    "Asian or Asian American",
    "Black or African American",
    "Hispanic, Latino, or Spanish origin",
    "Middle Eastern or Northern African",
    "Native American or Alaska Native",
    "White",
    "Another option not listed here or prefer not to answer"
]

# Create binary columns for each race in the master list
for race in race_master_list:
    df[race] = df['race'].apply(lambda x: 1 if race in x else 0)


# Drop the original
df.drop(columns = ['race'], inplace=True)

df.head()
```

```python
# Removing outliers from the dataset for annual_salary_usd and additional_compensation_usd using IQR method
Q1 = df['annual_salary_usd'].quantile(0.25)
Q3 = df['annual_salary_usd'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['annual_salary_usd'] >= (Q1 - 1.5 * IQR)) & (df['annual_salary_usd'] <= (Q3 + 1.5 * IQR))]
print(df.shape)

Q1 = df['additional_compensation_usd'].quantile(0.25)
Q3 = df['additional_compensation_usd'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['additional_compensation_usd'] >= (Q1 - 1.5 * IQR)) & (df['additional_compensation_usd'] <= (Q3 + 1.5 * IQR))]
print(df.shape)
```
